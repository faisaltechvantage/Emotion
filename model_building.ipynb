```python
import pandas as pd 
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
dataset= pd.read_csv("final_dataset.csv", index_col=0)
```

```python
# Split features and labels
X = dataset.drop(columns=['class'])
Y = dataset['class']
```

```python
Y
```

```python
X
```

```python
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
#Split data to train and test 
X_train, X_test, Y_train, Y_test = train_test_split (X,Y,test_size=0.2,random_state=42)
```

```python
pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Initialize SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)  # 'auto' balances all classes
```

```python
# Apply SMOTE to the training set
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, Y_train)
```

```python
# Optionally, check the new class distribution
print("Original class distribution:")
print(Y_train.value_counts())

print("Resampled class distribution:")
print(pd.Series(y_train_resampled).value_counts())
```

<!-- MLP Classifier -->

```python
#Create a Multi-Layer Classifier instance 
classifier = MLPClassifier(alpha=1.9,max_iter=700)
```

```python
#Fit the classifier
classifier.fit(X_train_resampled,y_train_resampled)
```

```python
#Calculate the score (Accuracy)
score = classifier.score(X_test,Y_test)
```

```python
print("score", score)
```

```python
print(classifier.classes_)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Evaluate the model
accuracy = accuracy_score(Y_test, y_pred)
report = classification_report(Y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```

```python
from joblib import dump

# Assuming 'model' is your MLP classifier
dump(classifier, 'models/mlp_classifier_model.joblib')
```
