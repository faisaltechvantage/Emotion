```python
# Import Libraries
import pandas as pd 
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
# Load DATA 
df = pd.read_csv('C:\EMOTION\Emotion\EmotionGithub\Classifiers\dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.label.unique()
```

```python
# Header for Features without Labels 
features = [str(i) for i in range(1,1583)]
```

```python
# Standarize the DATA 
X = df.loc[:,features].values

Y = df.loc[:,'label'].values
```

```python
X
```

```python
Y
```

```python
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
# PCA  
# n : Number of principal components 
n = 90
pca = PCA(n_components = n)
```

```python
pca
```

```python
X = pca.fit_transform(X)
```

```python
len(X[0])
```

```python
#Split data to train and test 
X_train, X_test, Y_train, Y_test = train_test_split (X,Y,test_size=0.2)
```

```python
#Create a Multi-Layer Classifier instance 
classifier = MLPClassifier(alpha=1.9,max_iter=700)
```

```python
#Fit the classifier
classifier.fit(X_train,Y_train)
```

```python
#Calculate the score (Accuracy)
score = classifier.score(X_test,Y_test)
```

```python
print(classifier.classes_)
```

```python
#Printing the score
print("score - test", score)
```

```python
score_train = classifier.score(X_train,Y_train)
```

```python
#Printing the score
print("score - train", score_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(Y_test, y_pred))
```

```python
# Classification result
print(classification_report(Y_test, y_pred))
```
