```python
import pandas as pd 
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load DATA 
df = pd.read_csv('Classifiers\dataset.csv')
df.head()
```

```python
df.label.unique()
```

```python
df.describe()
```

```python
df.shape
```

```python

# Header for Features without Labels 
features = [str(i) for i in range(1,1583)]
```

```python
# Standarize the DATA 
X = df.loc[:,features].values

Y = df.loc[:,'label'].values
```

```python
X = StandardScaler().fit_transform(X)

# PCA  
# n : Number of principal components 
n = 90
pca = PCA(n_components = n)

X = pca.fit_transform(X)
```

```python
#Split data to train and test 
X_train, X_test, Y_train, Y_test = train_test_split (X,Y,test_size=0.2)

#Create a Multi-Layer Classifier instance 
classifier = MLPClassifier(alpha=1.9,max_iter=700)
```

```python
#Fit the classifier
classifier.fit(X_train,Y_train)
```

```python
#Calculate the score (Accuracy) - Testing
score = classifier.score(X_test,Y_test)
```

```python
print(classifier.classes_)
```

```python
#Printing the score
print("Testing", score)
```

```python
#Calculate the score (Accuracy) - Training
score = classifier.score(X_train,Y_train)
```

```python
print("Training", score)
```

```python
from sklearn.metrics import classification_report

# Predict on the test set
Y_pred = classifier.predict(X_test)
```

```python
# Print classification report
print(classification_report(Y_test, Y_pred))
```
